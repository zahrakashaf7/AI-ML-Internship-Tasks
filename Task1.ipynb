# Task 1: Exploring and Visualizing the Iris Dataset

# Goal:
To explore and visualize the Iris flower dataset using Python libraries like pandas, matplotlib, and seaborn.  
We aim to understand data structure, feature relationships, and value distributions to draw initial insights.


# Step 1: Load the Dataset

We will use the `seaborn` library to load the built-in Iris dataset.  
This dataset contains measurements of 3 types of iris flowers: Setosa, Versicolor, and Virginica.


```python
# Import Libraries 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# iris dataset from seaborn
iris = sns.load_dataset("iris")   
```

# Step 2: Dataset Overview

We will check the structure of the dataset using basic pandas functions:
- `shape` to know the number of rows and columns
- `columns` to list feature names
- `head()` to preview the first few rows


```python
iris.head()  #head

print("Shape:", iris.shape) #Shape

print("Columns:", iris.columns.tolist()) #Columns
```

# Step 3: Summary Statistics

We use:
- `info()` to check data types and missing values
- `describe()` to view basic statistical summaries such as mean, min, max, and standard deviation


```python
iris.info()  #info

iris.describe() #describe
```

# Step 4: Scatter Plots

Scatter plots help us understand relationships between two numerical features.  
We'll plot:
- Sepal length vs Sepal width
- Petal length vs Petal width  
Each point is colored by the species of the flower.


```python
# Scatter Plot  Sepal
sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species")
plt.title("Sepal Length vs Sepal Width")
plt.show()

# Scatter Plot  Petal
sns.scatterplot(data=iris, x="petal_length", y="petal_width", hue="species")
plt.title("Petal Length vs Petal Width")
plt.show()
```

# Step 5: Histograms

Histograms show how values are distributed for each feature.  
We use this to check for:
- Normality
- Skewness
- Spread of data


```python
# Histogram
iris.hist(figsize=(10, 8))
plt.suptitle("Histogram of Iris Features")
plt.show()
```

# Step 6: Box Plots

Box plots help identify:
- Outliers
- Median values
- Interquartile range (IQR)

We'll compare features across the three species.


```python
# Box Plots
sns.boxplot(data=iris, x="species", y="sepal_length")
plt.title("Boxplot of Sepal Length by Species")
plt.show()
```

## Final Observations

- Petal features clearly separate the three species.
- Setosa flowers have shorter petal and sepal lengths.
- Box plots show Setosa has the smallest spread in features.
- No missing values were found in the dataset.
